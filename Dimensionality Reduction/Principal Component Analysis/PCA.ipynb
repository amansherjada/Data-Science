```python
x = df.iloc[:,:-1] #Features
y = df['class'] #Target
```


Principal Component Analysis (PCA) is an unsupervised machine learning technique used for dimensionality reduction. It aims to identify patterns in data by transforming the original variables into a new set of uncorrelated variables called principal components.

PCA does not require a target column because it focuses solely on the features (independent variables) present in the dataset. It works by identifying the directions (principal components) that capture the maximum variance in the data.

```python
x
```

### PCA

```python
sc = StandardScaler()
```

```python
x = sc.fit_transform(x)
```

```python
pd.DataFrame(x)
```

```python
pd.DataFrame(x).describe()
```

```python
x
```

```python
# from sklearn.decomposition import PCA
pca = PCA()
```

```python
pcs = pca.fit_transform(x)
```

```python
pd.DataFrame(pcs)
```

```python
pca.explained_variance_ratio_
```

- explained_variance_ratio_ Gives insight into how much information each principal component captures from the original data.
- The explained_variance_ratio_ tells you the proportion of variance (or information) captured by each principal component.- 
For example, if you have three principal components, and the explained_variance_ratio_ shows [0.7, 0.2, 0.1], it means the first component explains 70% of the variance, the second explains 20%, and the third explains 10%.

```python
np.cumsum(pca.explained_variance_ratio_)
```

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.title('Scree Plot', size =20)
plt.xlabel('Priciple Components')
plt.ylabel('Explained Varience')
plt.axvline(x = 1.0, color='red', linestyle='--')
plt.show()
```

```python
# First 2 princple componets is enough for us
```

```python
pca = PCA(n_components=2)
pcs = pca.fit_transform(x)
pcs
```

```python
pd.DataFrame(pcs)
```

```python
data = pd.DataFrame(pcs, columns=['PC1','PC2'])
data
```

```python
data['Class'] = y
# we stored the target coloum from original dataset into y (y = df['class'] #Target)
```

```python
data
```

```python
data['PC1'].mean()
```

```python
sns.scatterplot(x =data['PC1'], y = data['PC2'],s = 50, edgecolor = 'black', hue = data['Class'])
plt.scatter(pcs[:,0].mean(), pcs[:,1].mean(), s = 150, c= 'red' , edgecolor= 'black')
```

```python
data
```

```python
# Building Logistic Regression model oo the PCA generated data
```

```python
x = data.iloc[:,:-1]
y = data['Class']
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size=0.2, random_state=1)
```

```python
lg = LogisticRegression()
lg.fit(xtrain, ytrain)

ypred = lg.predict(xtest)
print(classification_report(ytest, ypred))
```

As we can see even after selecting the 50% data, still we are able to gte 97% accuracy of the model
