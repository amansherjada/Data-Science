```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split, KFold, LeaveOneOut,cross_val_score
from sklearn.metrics import classification_report
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/aishwaryamate/Datasets/main/fetal_health.csv')
df
```

## Train Test Split

```python
x = df.iloc[:,:-1]
y = df['fetal_health']
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2, random_state=45)
```

```python
dt = DecisionTreeClassifier()
dt.fit(xtrain, ytrain)
ypred = dt.predict(xtest)
```

```python
ypred
```

```python
print(classification_report(ytest, ypred))
```

## K-Fold Cross Vaildation

```python
#n_splits is number of folds
kf= KFold(n_splits=5)
# cross_val_score combine K-Fold Cross and Model 
# cv is Cross Validation
scores = cross_val_score(DecisionTreeClassifier(), x,y,cv=kf)
```

```python
scores
```

```python
#Overall Accuracy
scores.mean()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
kf = KFold(n_splits=5)

scores = cross_val_score(LogisticRegression(),x,y,cv=kf)
```

```python
scores
```

```python
#Overall Accuracy of Logistic Regression model
scores.mean()
```

## Leave One Out Cross Validation

```python
#Read the dataset
df = pd.read_csv('https://raw.githubusercontent.com/aishwaryamate/Datasets/main/Pima.csv',
                index_col=0)
df
```

```python
x = df.iloc[:,:-1]
y = df['class']
```

```python
lv_cv = LeaveOneOut()

lv_score = cross_val_score(DecisionTreeClassifier(),x,y, cv = lv_cv)
```

```python
lv_score
```

```python
# 1.0 implies perfect accuracy on a test fold (all predictions are correct).
# 0.0 indicates zero accuracy on a test fold (all predictions are incorrect).
```

```python
lv_score.mean()
```

```python
lv_score = cross_val_score(LogisticRegression(),x,y, cv = lv_cv)
```

```python
lv_score.mean()
```
