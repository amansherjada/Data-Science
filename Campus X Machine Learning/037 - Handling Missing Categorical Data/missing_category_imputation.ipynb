2 type for filling Categorical missing value

**1. Most Frequent Imputation {Mode}**
- Used when missing data is greater than 5%
- Used when data is Randomly Missing

**2. Missing Category Imputation (Create a New Catogory)**
- Used when missing data is greater than 5%
- Used when data is not Randomly Missing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/campusx-official/100-days-of-machine-learning/main/day37-handling-missing-categorical-data/train.csv',
                usecols=['GarageQual', 'FireplaceQu', 'SalePrice'])
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.isnull().mean()*100
```

```python
df['GarageQual'].value_counts().sort_values(ascending=False).plot.bar()
plt.xlabel('GarageQual')
plt.ylabel('Number of houses')
```

```python
df['GarageQual'].fillna('Missing', inplace=True)
```

```python
df['GarageQual'].value_counts().plot(kind='bar')
plt.xlabel('GarageQual')
plt.ylabel('No. of House')
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=['SalePrice']),df['SalePrice'],test_size=0.2)
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy='constant', fill_value='Missing')
```

```python
X_train= imputer.fit_transform(X_train)
X_test = imputer.fit_transform(X_test)
```
