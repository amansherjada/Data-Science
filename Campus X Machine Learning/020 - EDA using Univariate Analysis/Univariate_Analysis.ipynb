## Univariate Analysis

- Univariate analysis is a statistical method that involves the examination and interpretation of a single variable in isolation. In simpler terms, it focuses on analyzing the characteristics, properties, and patterns of one variable at a time. This type of analysis helps researchers and analysts understand the distribution, central tendency, and variability of the data within that single variable.

- Univariate analysis often includes descriptive statistics and graphical representations to summarize and visualize the key features of the variable under investigation. Common measures used in univariate analysis include mean, median, mode, range, variance, and standard deviation. Graphical representations may include histograms, bar charts, pie charts, and box plots to provide a visual summary of the variable's distribution.

- Overall, univariate analysis serves as a foundational step in the exploration and understanding of individual variables before moving on to more complex multivariate analyses that involve the simultaneous examination of multiple variables.

### 1. Categorical Columns EDA

**a. Countplot**

```python
import seaborn as sns
```

```python
sns.countplot(x = df['Survived'])
# 0 is Survived
# 1 is Not Survived
```

```python
df['Survived'].value_counts()
```

```python
df['Survived'].value_counts().plot(kind='bar')
```

```python
sns.countplot(x = df['Pclass'])
```

```python
sns.countplot(x = df['Sex'])
```

```python
sns.countplot(x = df['Embarked'])
```

**b. Pie Chart**

```python
import matplotlib.pyplot as plt
```

```python
df['Survived'].value_counts().plot(kind='pie', autopct='%1.1f',
                                  labels=['Survived', 'Not Survived'])
plt.title('Survival Distribution')
#plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
