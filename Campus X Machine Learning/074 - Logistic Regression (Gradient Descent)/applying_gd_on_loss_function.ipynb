### Why we apply Gradient descent on the loss function in logistic regression

In logistic regression, the goal is to find the optimal parameters (coefficients) that minimize the error in predicting the outcome of a binary classification problem. This optimization task is typically achieved by minimizing a loss function, commonly referred to as the logistic loss or cross-entropy loss.

The logistic loss function quantifies the difference between the predicted probabilities and the actual outcomes. Specifically, for binary classification, it measures how well the predicted probabilities align with the true binary labels.

Now, why do we apply gradient descent to minimize this loss function?

1. **Differentiable Function**: The logistic loss function is smooth and differentiable, which makes it amenable to optimization using gradient-based methods like gradient descent.

2. **Minimization Objective**: The objective of logistic regression is to minimize the logistic loss function. Gradient descent is an iterative optimization algorithm that moves in the direction of the steepest decrease of the loss function. By repeatedly updating the parameters in the direction that reduces the loss, gradient descent gradually converges to the optimal parameter values that minimize the loss.

3. **Efficiency**: For large datasets or high-dimensional feature spaces, computing the optimal parameters analytically may not be feasible or efficient. Gradient descent provides a computationally efficient way to find the optimal parameters iteratively.

4. **Flexibility**: Gradient descent is a general-purpose optimization algorithm that can be applied to various loss functions and models, making it widely applicable in machine learning.

In simple terms, we use gradient descent on the loss function in logistic regression because it helps us find the best "fit" for our model. Here's why:

1. **Minimizing Errors**: The loss function tells us how wrong our predictions are compared to the actual outcomes. By minimizing this loss function, we aim to make our predictions as accurate as possible.

2. **Gradient Descent**: Think of gradient descent as a step-by-step process of finding the lowest point (minimum) on a hill. In our case, the "hill" represents the loss function, and the lowest point is where the loss is smallest. We keep adjusting our parameters (coefficients) in the direction that decreases the loss until we reach that minimum.

3. **Efficiency**: Gradient descent is like taking small steps downhill. It's a smart and efficient way to find the best parameters, especially when dealing with large amounts of data or complex models.

```python
from sklearn.datasets import make_classification
import numpy as np
X, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=20)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
```

```python
from sklearn.linear_model import LogisticRegression
lor = LogisticRegression(penalty='none',solver='sag')
lor.fit(X,y)
```

In the code you provided, `penalty='none'` and `solver='sag'` are parameters used to configure the logistic regression model in scikit-learn.

- `penalty='none'`: This parameter specifies the type of regularization penalty to apply to the logistic regression model. Setting it to `'none'` means that no regularization penalty will be applied. Regularization penalties are typically used to prevent overfitting by penalizing large coefficients in the model. When `penalty` is set to `'none'`, the logistic regression model will be trained without any regularization.

- `solver='sag'`: This parameter specifies the optimization algorithm used to fit the logistic regression model. `'sag'` stands for Stochastic Average Gradient descent. It's an optimization algorithm that efficiently handles large datasets by using a stochastic gradient descent approach. It's particularly useful when the dataset is large and the number of features is high.

So, in summary, when you initialize the logistic regression model with `penalty='none'` and `solver='sag'`, you're configuring it to be trained without any regularization penalty and using the Stochastic Average Gradient descent algorithm for optimization.

1. `m1 = -(lor.coef_[0][0]/lor.coef_[0][1])`: This line calculates the slope of the decision boundary line. In logistic regression, the coefficients (`coef_`) represent the weights assigned to each feature. `lor.coef_[0][0]` represents the weight of the first feature, and `lor.coef_[0][1]` represents the weight of the second feature. So, `-(lor.coef_[0][0]/lor.coef_[0][1])` calculates the slope of the line by taking the negative ratio of the coefficient of the first feature to the coefficient of the second feature.

2. `b1 = -(lor.intercept_/lor.coef_[0][1])`: This line calculates the y-intercept of the decision boundary line. The intercept (`intercept_`) represents the bias term in the logistic regression model. `lor.intercept_` retrieves the intercept value, and `lor.coef_[0][1]` represents the weight of the second feature. So, `-(lor.intercept_/lor.coef_[0][1])` calculates the y-intercept by taking the negative ratio of the intercept to the coefficient of the second feature.

Together, `m1` and `b1` represent the slope and y-intercept of the decision boundary line, respectively, for the logistic regression model. These values can be used to plot the decision boundary line separating the classes in the feature space.

```python
def gd(X,y):
    
    X = np.insert(X,0,1,axis=1)
    weights = np.ones(X.shape[1])
    lr = 0.5 # Learning rate
    
    for i in range(5000):
        y_hat = sigmoid(np.dot(X,weights))
        weights = weights + lr*(np.dot((y-y_hat),X)/X.shape[0]) # (for this equation Refer Gradient_Descent_on_Loss_Function.pdf from the github repository)
        
    return weights[1:],weights[0]
```

This code defines a function named `gd` which implements the gradient descent algorithm to train a logistic regression model. Here's what the code does:

1. **Adding Bias Term**: The input features `X` are augmented with a bias term by inserting a column of ones as the first column. This is done to allow the model to learn a bias term (intercept) during training.

2. **Initializing Weights**: The weights vector is initialized with ones. This vector represents the coefficients of the logistic regression model.

3. **Learning Rate**: The learning rate (`lr`) is set to 0.5. This parameter controls the step size of each update during gradient descent.

4. **Gradient Descent Loop**: The code iterates through a loop for a fixed number of iterations (5000 in this case). In each iteration:
   - The predicted probabilities y_hat are computed using the sigmoid function applied to the dot product of input features and weights.
   - The weights are updated using the gradient descent update rule:

     `weights = weights + lr * (y - y_hat / n) X` (Refer Gradient_Descent_on_Loss_Function.pdf from the github repository)
     
     where \( y \) is the true labels, \( n \) is the number of samples, and \( X \) is the input feature matrix.
     
5. **Returning Weights**: After the loop, the function returns the learned weights corresponding to the coefficients of the logistic regression model. The bias term (intercept) is returned separately from the rest of the coefficients.

Overall, this function performs gradient descent to train a logistic regression model and returns the learned coefficients. The `sigmoid` function used within this function is assumed to be defined elsewhere in the code.

```python
def sigmoid(z):
    return 1/(1 + np.exp(-z))
```

This function `sigmoid(z)` implements the sigmoid activation function. The sigmoid function takes in a real-valued number `z` and maps it to a value between 0 and 1.

Here's a breakdown of how the function works:

1. **Input Parameter**: The function takes a single input parameter `z`, which could be a scalar, vector, or matrix.

2. **Sigmoid Transformation**: The function computes the sigmoid transformation of the input `z` using the formula:

   `sigmoid(z) = 1 / 1 + e^-z`
   
   where \( e \) is the base of the natural logarithm and \( e^{-z} \) is the exponential of the negative input.

3. **Output**: The output of the function is a value between 0 and 1, representing the probability of a binary outcome being true (or 1). Values close to 1 indicate high probability, while values close to 0 indicate low probability.

This sigmoid function is commonly used in logistic regression and other models to introduce non-linearity and squash the output to a range suitable for classification tasks.
