```python
# Converting to datetime datatype
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

```python
# 1. Extract year
date['date_year'] = date['date'].dt.year
date.head()
```

```python
# 1. Extract Month
date['date_month_no'] = date['date'].dt.month
date.head()
```

```python
# 1. Extract Month name
date['date_month_name'] = date['date'].dt.month_name()
date.head()
```

```python
# 1. Extract Day
date['date_day'] = date['date'].dt.day
date.head()
```

```python
# 1. Extract Day of week
date['date_dayofweek'] = date['date'].dt.dayofweek
date.head()
```

```python
# Extract Day of week name
date['date_day_name'] = date['date'].dt.day_name()
date.head()
```

np.where: This is a function from the NumPy library (np), which is used to conditionally assign values based on a condition. It has the syntax 

`np.where(condition, value_if_true, value_if_false)`

```python
# is weekend
date['date_is_weekend'] = np.where(date['date_day_name'].isin(['Sunday', 'Saturday']),
                                  1,0)
date.head()
```

```python
# Extract week of the year
date['date_week'] = date['date'].dt.isocalendar().week
date.head()
```

```python
# Extract Quarter
date['quarter'] = date['date'].dt.quarter
date.head()
```

```python
# Extract Semester
date['semester'] = np.where(date['quarter'].isin([1,2]), 1, 2)

date.head()
```

```python
# Extract Time elapsed between dates
import datetime

today = datetime.datetime.today()

today
```

```python

today - date['date']
```

```python
(today - date['date']).dt.days
```

```python

# Months passed

np.round((today -date['date']) / np.timedelta64(1, 'M'),0)
```
