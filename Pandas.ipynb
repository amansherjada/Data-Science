#### Pandas Filtering and Groupby

```python
dff.head()
```

```python
#Extract records where gender = Female
dff["gender"] == 'Female'
```

```python
dff[dff["gender"] == 'Female'] # this is the way 
```

```python
# Gender is female and salary is greater than 150000

# dff["gender"] == "Female"
# dff["salary"] > 150000

dff[(dff["gender"] == "Female") & (dff["salary"] > 150000)]
# and keyword is not allowed in pandas you have to write & for and operator
```

```python
dff[(dff["gender"] == "Female") | (dff["salary"] > 150000)]
# OR keyword is not allowed in pandas you have to write | for OR operator
```

```python
#What is the average salary of rank prof?
print("The Avg salary of Prof is: ",dff[dff["rank"]== "Prof"]["salary"].mean())
```

```python
# Average Salary
dff['salary'].mean()
```

```python
#Average salary of prof rank and gender is Male
dff[(dff["rank"]== "Prof") & (dff["gender"]=="Male")] ["salary"].mean()
```

```python
#Average salary of prof rank and gender is Female
dff[(dff['rank'] == "Prof") & (dff['gender'] == 'Female')]['salary'].mean()
```

```python
#Grouby aggregate method
```

```python
#Average salary of rank 
dff.groupby("rank")["salary"].mean()
```

```python
dff.groupby("rank")["salary"].agg({'mean','median','min','max'})
```

```python
ca = pd.read_csv('https://raw.githubusercontent.com/aishwaryamate/Datasets/main/nba.csv')
ca
```

```python
# Q1 Extract the data where Team is boston celtics and Weight is less than 100
# Team is boston celtics
# ca["Team"] == "Boston Celtics"
# Weight is less than 100
# ca["Weight"] < 100
# Solution:
ca[(ca["Team"] == "Boston Celtics") & (ca["Weight"] > 100)]
```

```python
# Q2. What is the avearge salary of players of team 'Los Angeles Lakers'

# Solution:
print("Avearge salary of players of team 'Los Angeles Lakers' is:", ca[ca["Team"] == "Los Angeles Lakers"]["Salary"].mean())
```

```python
# Q3. What is the average salary of player based of various position?
#Solution:
ca.groupby('Position')['Salary'].mean()
```

```python
# Q3. What is the average salary of player based of various position?
#ChatGPT Solution:
average_salary_by_position = ca.groupby('Position')['Salary'].mean()
result_df = average_salary_by_position.reset_index()
result_df.columns = ['Position', 'AverageSalary']
print(result_df)
```

```python
# Q4. Which players belong to San Diego State college.
#Solution 1:
belong =ca[ca["College"] == "San Diego State"]
belong[["Name", "College"]]
```

```python
# Q4.
# Solution 2: Changed the name of college name to check 
ca[ca["College"] == "North Carolina"][["Name","College"]]
```

```python
#Q5. Which player from Toronto Raptors team has the highest salary?

# Solution:
ca[ca["Team"] == "Toronto Raptors"]["Salary"].max()
```
